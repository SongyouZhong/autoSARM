```python
''' Just simply search for a single fragments 
such as N(CC1)CCC1c2nc3cc[nH]c3cc2
'''
from rdkit import Chem, DataStructs
import pandas as pd
import numpy as np
from rdkit.Chem.Scaffolds.MurckoScaffold import GetScaffoldForMol,MakeScaffoldGeneric,MurckoScaffoldSmiles,MurckoScaffoldSmilesFromSmiles
from utils.common_utils import mapper,csvToExcel,get_mol,float_row,sort_mol_sim_df
from utils.sarm_utils import fragmize_on_bonds, get_complete_frag, get_core
from rdkit import Chem, DataStructs

sub_smarts = "CC(c1c(cc(C2CC[NH]CC2)cc3)c3[nH]c1c4cn5ncnc5cc4)C" #"CC(C1=C(C)NC2=C(C)C(C)=CN=C12)C"      
submol_smarts = Chem.MolFromSmarts(sub_smarts)
df_act=pd.read_csv("Data/MedChemAct_Cano_Sele_JAN16.csv")
df_act=df_act.dropna(subset=["TLR8to7","Affinity_TLR7"])
df_act=float_row(df_act, cols=["TLR8to7","Affinity_TLR7"])
# df_act=df_act[df_act["TLR8to7"]>0]
# df_act=df_act[df_act["Affinity_TLR7"]<100]
matched_list=[]
for ismi in df_act["Cano_SMILES"].to_list():
    # print(ismi)
    imol=Chem.MolFromSmiles(ismi)
    if imol==None: continue
    matched=imol.GetSubstructMatches(submol_smarts)
    if len(matched)>0:
        matched_list.append(ismi)
        print(f"Matched: {ismi}")    
df_act['index']=df_act['Cano_SMILES'] 
df_act=df_act.set_index("index")
df_matched=df_act.loc[matched_list]
df_matched=sort_mol_sim_df(df_matched, col='Cano_SMILES')
df_matched.to_csv(f'Results/matched_{sub_smarts}.csv',index=True)
csvToExcel(f'Results/matched_{sub_smarts}.csv', imgCols=['SMILES'],save_file=f'Results/matched_{sub_smarts}.xlsx',max_imgs=500,column=False)
print("Total matched:", len(df_matched))
print("Matched examples:", df_matched[:5])
```

```python
'''  Functions for matching for a key fragment (Single cut)
For example:  AAAAA-X; First the high selective dataset was searched, then similar frags of high selective set was searched from the larger active set.
'''
from rdkit import Chem, DataStructs
from rdkit.Chem import AllChem,Draw
from rdkit.Chem import MCS
import pandas as pd
import numpy as np
import copy,re
from rdkit.Chem.Scaffolds.MurckoScaffold import GetScaffoldForMol,MakeScaffoldGeneric,MurckoScaffoldSmiles,MurckoScaffoldSmilesFromSmiles
from utils.common_utils import mapper,csvToExcel,get_mol,float_row,sort_mol_sim_df
from utils.sarm_utils import fragmize_on_bonds, get_complete_frag, get_core
import os,sys
from IPython.display import display, SVG,display_svg


frag_smi="N(CC1)CCC1c2nc3cc([*])[nH]c3cc2"  ## this can be transformed into mol
df_sele=pd.read_csv("Data/seleIdx20TLR820.csv")
df_sele=sort_mol_sim_df(df_sele, col='Cano_SMILES')
frag_smarts=["N(CC1)CCC1a2aa3aa([*])[nH]a3aa2"] ## these can extend the match domain

matched_list=[]
ordered_frags_list=[]
for ismi in df_sele["Cano_SMILES"].to_list():
    imol=Chem.MolFromSmiles(ismi)
    if imol==None: continue
    for ismarts in frag_smarts:
        ismarts_mol = Chem.MolFromSmarts(ismarts)
        matched=imol.GetSubstructMatches(ismarts_mol)
        if len(matched)>0:
            ordered_frags=fragmize_on_bonds(ismi, ismarts, frag_smi)
            print('ordered_frags= ',ordered_frags)
            ordered_frags_list.append(ordered_frags)
            matched_list.append(ismi)
            
df_sele['index']=df_sele['Cano_SMILES'] 
df_sele=df_sele.set_index("index")
df_matched=pd.DataFrame(df_sele.loc[matched_list])
df_matched[['Frag','Core']]=ordered_frags_list

df_act=pd.read_csv("Selectivity_Jan16/MedChemAct_Cano_Sele_JAN16.csv")
df_act=df_act.dropna(subset=["TLR8to7","Affinity_TLR7"])
df_act=float_row(df_act, cols=["TLR8to7","Affinity_TLR7"])
results_match_list=[]
group_idx=-1  ## index of the reference molecules
explored_frags=[]
for idx,irow in df_matched.iterrows():
    ''' Selected high selective database. '''
    ifrag=irow['Frag']
    if ifrag in explored_frags:
        continue
    else:
        explored_frags.append(ifrag)
    icore=irow['Core']
    group_idx+=1
    re_p=re.compile(r'\(\*\)|\*|\[\*\]')
    ifrag_nodummy = re.sub(re_p, '', ifrag)  ## remove dummy atoms
    ifrag_mol = Chem.MolFromSmarts(ifrag_nodummy)
    results_match_list.append(irow[['SMILES','Affinity_TLR7','Affinity_TLR8','Affinity_TLR9','Cano_SMILES','TLR8to7']].to_list()+[group_idx,icore])
    for jdx, jrow in df_act.iterrows():
        ''' The whole active database. '''
        jmol=Chem.MolFromSmiles(jrow['Cano_SMILES'])
        matched=jmol.GetSubstructMatches(ifrag_mol)
        if len(matched)>0:
            jfrag_complete = get_complete_frag(jmol, ifrag)[0]
            # print(ifrag, jfrag_complete)
            # sys.exit()
            if jfrag_complete!=ifrag: continue
            jcore = get_core(jmol, [ifrag])
            results_match_list.append(jrow[['SMILES','Affinity_TLR7','Affinity_TLR8','Affinity_TLR9','Cano_SMILES','TLR8to7']].to_list()+[group_idx,jcore])
            print(jrow['Cano_SMILES'])
                
df_res_match=pd.DataFrame(results_match_list, columns=['SMILES','Affinity_TLR7','Affinity_TLR8','Affinity_TLR9','Cano_SMILES','TLR8to7','Group','Frag'])    
df_res_match=df_res_match.drop_duplicates('Cano_SMILES') 
d_count=pd.DataFrame(df_res_match["Group"].value_counts())   
d_count.columns=['Count']
d_count=d_count[d_count['Count']==1]
del_idxes=d_count.index
df_res_match=df_res_match[~df_res_match['Group'].isin(del_idxes)]
df_res_match=pd.DataFrame(df_res_match)
ordered_index=[]
group_list=list(set(df_res_match['Group']))
for igrp in group_list:
    tmp_df=df_res_match[df_res_match['Group']==igrp]
    tmp_df=sort_mol_sim_df(tmp_df, col='Cano_SMILES')
    ordered_index.extend(tmp_df.index)
df_res_match=df_res_match.loc[ordered_index]    
df_res_match.to_csv("Results/seleIdx20TLR820_CoreA.csv", index=None)
```
